```python
import pandas as pd
import re
```

```python
text         = pd.read_csv("../data/resultados/sem_siglas/resultado_sem_siglas", sep = ";")
scores       = pd.read_csv("../data/resultados/sem_siglas/resultado_sem_siglas_scores.csv")
prevalencias = pd.read_csv("../data/prevalencia_e_scores_com_siglas.csv")
classes      = prevalencias["Diagnostico"]
```

```python
def calcularIntervalo(id, text):
    intervalo = text[" resultado"][id]
    intervalo = intervalo[1:-1]
    intervalo = intervalo.split("[")
    del intervalo[0]
    intervalo = re.sub('],','', intervalo[numClasse])
    intervalo = re.sub(']','', intervalo)
    intervalo = re.sub(' ','', intervalo)
    intervalo = [int(x) for x in intervalo.split(",")]
    ini       = max(0, intervalo[0] - 20)
    fim       = min(intervalo[1] + 20, len(text[" texto"][id]))
    intervalo = text[" texto"][id][ini:fim]
    return intervalo
```

```python
numClasse   = 1
qtdEsperada = len(scores) * prevalencias["Prevalencia"][numClasse]
classeAtual = classes[numClasse]
print()
notaDeCorte = 90
slices = {}

#Seleciona os ids que atingem a nota de corte
resultTotal = len(scores[scores[classeAtual] >= notaDeCorte].index.values.astype(int))
result      = scores[scores[classeAtual].between(notaDeCorte, notaDeCorte+10)].index.values.astype(int)


for id in result:
    intervalo = calcularIntervalo(id, text)
    if intervalo not in slices:
        slices[intervalo] = 1
    else:
        slices[intervalo] += 1
        
slices = {k: v for k, v in sorted(slices.items(), reverse = True, key=lambda item: item[1])}
        
print("Classe sendo analisada: ", classeAtual)
print()
print("Quantidade Esperada:", qtdEsperada)
print()
print("Quantidade Encontrada:", resultTotal)
print()
print("Mostrando",len(slices),"resultados únicos. \n")
display(slices)
```
