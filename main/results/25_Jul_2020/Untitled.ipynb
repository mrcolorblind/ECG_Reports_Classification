```python
import pandas as pd
import numpy as np
```

```python
r1 = pd.read_csv("result_25_Jul.csv")
r2 = pd.read_csv("results_final.csv")
```

```python
float(r1[r1["Class"] == r2["Class"][0]]["Recall"])
```

```python
db = []
for i in range(35):
    row = []
    row.append(r2["Class"][i])
    row.append(float(r1[r1["Class"] == r2["Class"][i]]["Precision"]))
    row.append(r2["Precision_lda"][i])
    row.append(r2["Precision_regex"][i])
    row.append(float(r1[r1["Class"] == r2["Class"][i]]["Recall"]))
    row.append(r2["Recall_lda"][i])
    row.append(r2["Recall_regex"][i])
    row.append(float(r1[r1["Class"] == r2["Class"][i]]["F1"]))
    row.append(r2["F1_lda"][i])
    row.append(r2["F1_regex"][i])
    row.append(r2["Ocurrences"][i])
    db.append(row)
```

```python
db = np.array(db)
df = {}
df["Class"] = list(db[:,0])
df["Precision_PM"] = list(db[:,1])
df["Precision_LDA"] = list(db[:,2])
df["Precision_Regex"] = list(db[:,3])
df["Recall_PM"] = list(db[:,4])
df["Recall_LDA"] = list(db[:,5])
df["Recall_Regex"] = list(db[:,6])
df["F1_PM"] = list(db[:,7])
df["F1_LDA"] = list(db[:,8])
df["F1_Regex"] = list(db[:,9])
df["Ocurrences"] = list(db[:,10])
```

```python
pd.DataFrame(df).to_csv("results_final_complete.csv")
```
